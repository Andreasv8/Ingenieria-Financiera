Obtén datos con yfinance desde 2022 de netflix y otras 4 empresas (tú decide)

```python
import yfinance as yf
data = yf.download(["NFLX", "MSFT", "META", "NVDA", "DIS"], start="2022-01-01", end="2025-01-26")["Close"]
data
```

Graficar los precios de cierre

```python
import matplotlib.pyplot as plt

plt.plot(data.index, data[["NFLX", "MSFT", "META", "NVDA", "DIS"]], label="ACCIONES")
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Precio de las Acciones a lo largo del tiempo')

plt.legend()
plt.show()
```

Si hubieras invertido 100 dolares en cada empresa al inicio (1/1/2022) , cuánto valdría tu inversión hoy para cada empresa? Asume que puedes comprar fracciones de acciones.

```python
primerdato_netflix = data["NFLX"].iloc[0]
primerdato_netflix
```

```python
rendimiento_netflix = 100/primerdato_netflix
rendimiento_netflix
```

```python
ultimodato_netflix = data["NFLX"].iloc[-1]
ultimodato_netflix
```

```python
inversion_netflix = rendimiento_netflix * ultimodato_netflix
inversion_netflix
```
