(5 pts) ¿Cuál es la diferencia entre una regresión lineal simple y una regresión lineal múltiple? Da un ejemplo de cada una.

La diferencia entre ambas regresiones es el número de variables que utilizamos para hacer la predicción. En una regresión lineal simple únicamente se utiliza un varible. Por ejemplo, si queremos predecir cuantos helados se compran en una heladeria en esta regresión lineal solo utilizaremos la estación del año (invierno, verano) si queremos que la regresión sea más precisa recurrimos a un regresión lineal multiple. Ahora utilizaremos no solo la variable de estación del año, también utilizaremos temperatura, día de la semana, si es día de promoción o no, etc. (más de una variable).

(5 pts) Explica el problema de overfitting y cómo podrías mitigarlo.

Cuando decimos que hay overfitting lo que queremos decir es que hicimos que el modelo se entrenara demasiado, es decir que es ahora muy dependiente de los datos que le dimos y puede qeu cuando se le presenten nuevos datos la predicción no sea la mejor. Esto lo mitigamos haciendo uso del método cross validation. Este método basicamente nos ayuda a probar que tan bueno es el modelo prediciendo utilizando los datos que no fueron utilizados en el train. Así podemos ver si realmente es un buen modelo o se hizo muy dependiente a los datos utilizados en el train.

(5 pts) En regresión polinomial, ¿por qué es necesario realizar un análisis de validación cruzada antes de elegir el grado del polinomio?

Debemos de hacer este análisis de validación cruzada porque al momento de elegir el grado del polinomio estamos eligiendo que tan curva es la curva que utilizamos para hacer la predicción. El grado de polinomio dependiendo que tan pequeño o grande sea nos puede hacer un modelo demasiaso dependiente de los datos de train o el contrario, por eso al hacer el cross validation tenemos una mejor idea de esto para así poder elegir el grado correcto.

(5 pts) ¿Qué es una prueba de hipótesis en el contexto de regresión lineal? Explica cómo se interpreta el p-value de un coeficiente.

Cuando hacemos pruebas de hipótesis en regresión lineal lo que realmente se trata de comprobar es si la varible que estamos utilizando para nuestra predicción es significante o no. La hipótesis nula sería que era varible es igual a cero es decir no es significativa para la predicción. Por otro lado la hipótesis alterna sería que es diferente de cero (si es significativa). Al momento de analizar el p-value debemos de considerar los siguientes parámetros. Si el p-value es menor a 0.05 significa que la varible es significativo y si es mayor entonces no lo es. De esta manera es como podemos saber si esa variable nos sirve o no para hacer nuestras predicciones.

De esta manera podemos tomar en cuenta estos datos y utilizarlos a nuestro favor. En clase por ejemplo cuando teníamos muchas categorias dentro de una variable categórica entonces optabamos por utilizar el método One hot encoding donde se utiliza el promedio (no siempre es lo optimo) esto se hace para hacer el proceso más rápido solo en caso de tener muchos grupos en una variable categorica.

(20 pts) Imagina que entrenas un modelo de regresión polinomial de grado 5 y obtienes un 𝑅2 de 0.98 en el conjunto de entrenamiento y 0.65 en el conjunto de prueba.

- a) ¿Qué problema podrías estar enfrentando?

teniendo un R2 tan alto en el entrenamiento nos puede indicar que hay un problema de overfitting.

- b) ¿Cómo lo solucionarías sin reducir demasiado la capacidad del modelo?

Podríamos utilizar la regresión Ridge o Lasso para hacer esto. El método Lasso elimina esas variables que no son muy necesarias para la predicción. Y el método Ridge modifica el peso de las variables para que una no sea la que más peso tengo y este se distribuya.


`````python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
`````

`````python
df = pd.DataFrame({
    "Tamaño_m2": [50, 80, 120, 200, 150, 90, 175, 60, 220, 130],
    "Habitaciones": [1, 2, 3, 4, 3, 2, 3, 1, 5, 3],
    "Precio_1000s": [110, 170, 250, 400, 270, 200, 330, 190, 600, 380]  })
`````

`````python
df = df[df.Precio_1000s != df.Precio_1000s.max()]
`````

`````python
y = df['Precio_1000s']
X = df[['Tamaño_m2']]
model = linear_model.LinearRegression().fit(X, y)
predictions = model.predict(X)
r2_score(y_true=y, y_pred=predictions)
`````

`````python
y = df['Precio_1000s']
X = df[['Habitaciones']]
model = linear_model.LinearRegression().fit(X, y)
predictions = model.predict(X)
r2_score(y_true=y, y_pred=predictions)
`````

`````python
y = df['Precio_1000s']
X = df[['Tamaño_m2', 'Habitaciones']]
model = linear_model.LinearRegression().fit(X, y)
predictions = model.predict(X)
r2_score(y_true=y, y_pred=predictions)
`````

El R^2 utilizando ambas variables es de 0.8145, podemos ver que esto es mejor que si solo utilizamos una de las variables. Es mejor utilizar ambas.

`````python
# Interepto
model.intercept_
`````

`````python
# Coeficientes
model.coef_
`````

Los coeficientes son 1.07 y 34.33
