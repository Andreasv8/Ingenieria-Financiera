## **Sección 1: Regresión Logística** (30 puntos)  

1. **(10 pts)** Explica la diferencia entre la regresión logística **lineal** y la **polinomial**. ¿En qué casos es recomendable usar la versión polinomial?  

La principal diferencia entre una regresión logística lineal y una polinomial es que la lineal utiliza únicamente una relación lineal entre las variables dependientes y la independiente para predecir. Por otro lado, la polinomial utiliza también otros términos, por ejemplo cuadrados o cubicos para hacer la predicción.
Se utiliza o se recomienda utilizar la polinomial cuando la relación de la variable objetivo no es lineal con las otras variables. Por ejemplo, que una variable "y" dependa de otra variable "y". Si vemos la gráfica de dispersión de los datos y la linea no es recta o hay algún tipo de corvatura, se recomendaría utilizar una polinomial.

2. **(10 pts)** Explica como mediante decenso en gradiente y maxima verosimilitud creamos una regresión lógisitica  

Para una regresión logistica se deben de cumplir ciertos pasos, primero se debe de buscar maximizar la verosimilitud. La verosimilitud en palabras simples es que tan probable es que la predicción con la regresión sea cercana a los datos reales. Así podemos saber si nuestro modelo es un buen modelo o no. Ya que se máximiza la función de verosimilitud ahora debemos tambien buscar minimizar el log-loss, que básicamente es la diferencia entre las probabilidades predichas y las reales. Con decenso en gradiente podemos ir ajustando los parámetros y ver cuales son los que minimizan esta función de perdida y cuales maximizan la función de verosimilitud. Así el modelo de regresión logistica esta lo mejor entrenado posible.

Los odds se utilizan para saber cual es la probabilidad de que una probabilidad suceda. ## **Sección 5: Redes Neuronales y Perceptrón Multicapa** (20 puntos)  

13. **(5 pts)** Explica que es una red neuronal, que hace, como funciona, etc.   

Una red nueronal es un modelo que aprende patrones dificiles y complejos para poder hacer diferentes predicciones. Tiene diferentes capas, la capa de entrada que es los datos en crudo, las capas ocultas, y la capa de salida que sería el resultado que buscamos. La red neuronal se puede entrenar a nuestro gusto con los parámetrsi que pongamos en las capas.

14. **(5 pts)** ¿Cuál es el propósito de la **backpropagation** en el entrenamiento de redes neuronales?  

El proposito es que la red neuronal pueda aprender de los errores anteriores que ha cometido y lo hace ajustando los parámetros para poder disminuir la función de perdida y mejorar su capacidad para predecir.


15. **(5 pts)** A grandes rasgos, explica como obtenemos los coeficientes de una red neuronal  

Primero los pesos se asignan de manera aletoria, después de hace el forward propagation en cada capa. Luego tenemos que calcular el error es decir lo que de predijo y la salida real. Luego hacemos backpropagation para que la red calcule que pesos reducir o modificar para reducir el error. Después se hace el cambio de los pesos y se entrena varias veces para al final poder obtener los mejores coeficientes que minimicen el error y maximicen el poder predictivo.

16. **(5 pts)** Explica que es softmax, para que sirve y como se calcula

El softmax le da una calificación de que tan probable es que suceda cada clase. A comparación de la regresión logística aquí no es binario y se puede utilizar con más de 2 clases. Básicamente toma los valores de cada clase, calcula sus exponenciales y los suma todos, depúes cada exponencial de cada clase lo divide entre la suma total y así calcula las probabilidades de cada clase.
