## Sección I: Árboles de decisión (15 puntos)

### 1. (5 pts)  
Explica cómo se construye un árbol y qué criterio usa para decidir los splits. Explica tanto para el caso de clasificación como de regresión.

Para contruir un árbol de decisión se utiliza una serie de escaloes que busca encontrar los puntos de corte donde de redusca la varianza o la impureza, esto para llegar a la mejor predicción posible o el grupo más "puro" posible.

En el caso de la regresión: para elegir los splits
Primero se debe de decidir que carácteristica utilizar (por ejemplo valor promedio de una casa) para nuestra regresión y luego ordenar los datos de menor a mayor. Despúes se puede intentar hacer cortes en puntos medios entre los datos. En cada posible corte, hacemos una división y se hacen los grupos izquierdo(menor que) y derecho(mayor que), al decir mayor y menor que me refiero al valor de corte en esa división. Luego utilizando la fromula de reducción de varianza, calculamos en cada corte y buscamos el que más maximicé esa función. Ya que nuestro objetivo es reducir la varianza lo más posible. Este proceso se repite hasta que ya no tenga sentido seguir dividiendo más.

Ahora para el caso de clasificación:
Los pasos iniciales son similares, elegir la carácteristica y acomodar los datos, probar divisiones y separar en grupos (izq y derecho). Aquí la diferencia es que no buscamos reducir la varianza, si no que buscamos reducir la impureza, y esto lo hacemos con el indice de Gini o Entropía. Después de calcular la reducción de impureza se elige el que maximixé más la reducción de impureza y se repite el proceso de nuevo.

El sobreajuste llega a pasar cuando el arból se apoya demasiado a los datos de entrenamiento y comienza a hacer malas predicciones, por ejemplo si tienes un data set de casas que están a la venta y resulta que en una zona que es cara hay una casa que cuesta 6M y tiene la caracteristica de 3 habitaciones, al momento en el que otra casa en diferente zona con la misma cantidad de habitaciones entre al dataset se va a predecir que va a valer alrededor de los 6M pero la zona cambia mucho el precio y se hace un overfitting porque el modelo ya apredio que 3 habitaciones = 6M. Para evitar esto usamos la función max.depht para limitar la profundidad del árbol, o la función min_samples_split para evitar que el árbol haga divisiones en grupos muy pequeños que estadisticamente popdrían reducirse a cero.

## Sección III: Gradient Boosting (25 puntos)

### 6. (10 pts)  
Explica, paso a paso, cómo funciona el algoritmo de **Gradient Boosting**. Incluye el concepto de residuales y cómo se minimiza la pérdida en cada iteración.

El primer paso para hacer Gradient Boosting es hacer una predicción con el promerio de los valores reales con los que ya contamos, luego al obtener los resultados de esta predicción se calcular los residuales. Los residuales son simplemente la diferencia entre el valor real y el valor de la predicción. Luego se entrena un nuevo árbol ya que se reconoce el error del árbol anterior y esto lo cambia para el siguiente, y el modelo se va mejorando aprendiendo de esos errores anteriores. Para medir aquí utilizamos los parametros de MSE o el de log-loss.
